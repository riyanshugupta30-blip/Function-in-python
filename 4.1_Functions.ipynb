```python
def even_num(num):
    """This functions is find even or odd"""
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
even_num(4)
print(even_num.__doc__)
```

```python
## Default parameters
def greek(name="guest"):
    print(f"Hello {name}")
greek()
```

```python
# *args and **kwargs
def print_number(*args):
    for item in args:
        print(f"{item}")
print_number(1,2,3,4,5,6,7,8,"riyanshu")
```

```python
def func1(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
func1(name="Riyasnhu",age=24,city="varanasi")
```

```python
def print_details(*args,**kwargs):
    for item in args:
        print(f"Positional argument{item}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_details(1,2,3,"Riyasnhu",name="Riyasnhu",age=24,city="varanasi")
```

```python
## Return statemen
def multiply(a,b):
    return a*b
multiply(2,3)
```
