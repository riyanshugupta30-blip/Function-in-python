#### Lambda function in python
Lambda functions are small anonymous functions defined using the lambda keyword.they can have any number of argument
but only one expression. They are commenly used for short opretions or as argument to higher-order functions.



```python
# Syntex
lambda arguments: expression
```

```python
addition=lambda a,b:a+b
print(type(addition))
print(addition(4,5))
```

```python
even=lambda num:num%2==0
print(even(8))
```

```python
addition1=lambda s,y,z:s+y+z
print(addition1(4,5,7))
```

```python
number=[1,2,3,4,5,6]
list(map(lambda x:x**2,number))
```
