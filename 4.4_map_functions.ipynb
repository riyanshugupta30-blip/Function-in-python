#### The map() function in python
the map() function applies a given function to all items in an input list (or any other iterable) and returns a map
object(an interator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x
number=[1,2,3,4,5,6,7,8]
list(map(square,number))
map()
```

```python
## map() multiple iterables
number1=[1,2,3]
number2=[4,5,6]
added_number=list(map(lambda x,y:x+y,number1,number2))
print(added_number)
```

```python
## map() to convert a list of string go to integers
##use map to convert strings to integers
str_number=['1','2','3','4','5']
int_number=list(map(int,str_number))
print(int_number)
```

```python
words=["apple","banana","cherry"]
upper_words=list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']
people=[
    {"name":"Riyanshu","age":21},
    {"name":"Rocky","age":34}
]
list(map(get_name,people))

```
